```python
import pandas as pd
from sklearn.cluster import KMeans
```

```python
import matplotlib.pyplot as plt
```

```python
# Cargar datos desde un archivo CSV
df = pd.read_excel('transporte3.xlsx')
```

```python
# Seleccionamos las características relevantes para el clustering
features = ['Pasajeros', 'Pasajerosdiatipicolaboral', 'Pasajerosdiasabado', 'Pasajerosdiafestivo', 'DensidadPoblacion', 'Congestion2']

# Preparamos los datos para el clustering
df_clustering = df[features]

# Creamos una instancia del modelo KMeans con 3 clusters
num_clusters = 3
kmeans = KMeans(n_clusters=num_clusters, random_state=42)

# Entrenamos el modelo con nuestros datos
kmeans.fit(df_clustering)

# Obtenemos las etiquetas de los clusters para cada muestra
labels = kmeans.labels_

# Añadimos las etiquetas de los clusters al DataFrame original
df['Cluster'] = labels

# Visualizamos los resultados
plt.figure(figsize=(10, 6))
for cluster in range(num_clusters):
    cluster_data = df[df['Cluster'] == cluster]
    plt.scatter(cluster_data['Ciudad'], cluster_data['Pasajeros'], label=f'Cluster {cluster}')
plt.xlabel('Ciudad')
plt.ylabel('Pasajeros')
plt.title('Clusters de datos de transporte masivo por ciudad')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
